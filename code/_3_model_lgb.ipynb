```python
import pandas as pd
```

```python
import lightgbm as lgb
```

```python
import gc
```

```python
train = pd.read_pickle("../data/merge_data/final_train.pickle").reset_index()
train = train.set_index("instance_id")
val = pd.read_pickle("../data/merge_data/final_val.pickle").reset_index()
val = val.set_index("instance_id")
test = pd.read_pickle("../data/merge_data/final_test.pickle").reset_index()
test = test.set_index("instance_id")
```

```python
del train
del val
gc.collect()
```

```python
drop_cols = ['is_train',
             'shop_review_num_level_0',
             'shop_review_num_level_1',
             'shop_star_level_4999',
            ]
train = train.drop(drop_cols,axis=1)
test = test.drop(drop_cols,axis=1)
val = val.drop(drop_cols,axis=1)
```

```python
X_tr = train.drop(["is_trade_x"],axis=1)
y_tr = train["is_trade_x"]
X_val = val.drop(["is_trade_x"],axis=1)
y_val = val["is_trade_x"]
```

```python
X_tr = X_tr.drop(["predict_category_property"],axis=1)
```

```python
X_val = X_val.drop(["predict_category_property"],axis=1)
```

```python
X_val.columns
```

```python
drop_cols = []
for i in range(5000,5021):
    drop = "shop_star_level_%d"%(i)
    drop_cols.append(drop)
for i in range(3000,3011):
    drop = "user_star_level_%d"%(i)
    drop_cols.append(drop)
```

```python
drop_cols
```

```python
X_tr = X_tr.drop(drop_cols,axis=1)
X_val = X_val.drop(drop_cols,axis=1)
```

```python
X_tr.shape
```

```python
model = lgb.LGBMClassifier(boosting_type='gbdt', num_leaves=29, max_depth=3, learning_rate=0.01, n_estimators=10000,
                           max_bin=425, subsample_for_bin=50000, objective='binary', min_split_gain=0,
                           min_child_weight=5, min_child_samples=10, subsample=1, subsample_freq=1,
                           colsample_bytree=1, reg_alpha=3, reg_lambda=5, seed=1000, nthread=-1, silent=True)
```

```python
model.fit(X_tr, y_tr, eval_metric='logloss',eval_set=[(X_tr, y_tr), (X_val, y_val)],early_stopping_rounds=100)
gc.collect()
```

```python
y_pred = model.predict(X_val)
```

```python
pd.Series(y_pred).value_counts()
```

```python
drop_columns = ["predict_category_property"] + drop_cols
```

```python
test = test.drop(drop_columns,axis=1)
```

```python
set(test.columns) - set(X_tr.columns)
```

```python
test = test.drop(["day"],axis=1)
```

```python
proba = model.predict_proba(test)
```

```python
result_v1 = pd.DataFrame(test.index)
result_v1["predicted_score"] = proba[:,-1]
```

```python
result_v1.to_csv("../data/model/lgb_preds.txt",sep=" ",index=False,line_terminator="\r")
```
